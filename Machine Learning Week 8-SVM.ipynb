```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_wine

wine = load_wine()
feature = pd.DataFrame(wine['data'],columns=wine['feature_names'])
target = pd.DataFrame(wine['target'],columns=['class'])

data = pd.concat([feature,target],axis=1)
data
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('ggplot')
g=sns.FacetGrid(data,hue='class',size=5)
g.map(plt.scatter,"alcohol" , "od280/od315_of_diluted_wines")
g.add_legend()
```

```python
from sklearn.model_selection import train_test_split
x=data.iloc[:,[0,11]].values
y=data.iloc[:,-1].values
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=1,stratify=y)
```

```python
from sklearn.svm import SVC
svm = SVC(kernel='liner')
svm_rbf = SVC(kernel='rbf',gamma='scale')
svm_poly = SVC(kernel='poly',degree=3)
```

```python
from sklearn.metrics import accuracy_score

svm=svm.fit(x_train,y_train)
y_train_pred=svm.predict(x_train)
y_test_pred=svm.predict(x_test)
svm_train=accuracy_score(y_train,y_train_pred)
svm_test=accuracy_score(y_test,y_test_pred)
print('decision tree train/test accutancies %.3f/%.3f'%(svm_train,svm_test))
```
