```python
#2.(3)畫出g'(Θ0)的函數圖
import numpy as np
import matplotlib.pyplot as plt

x=np.linspace(-3,3)
y=x*5-2
plt.plot(x,x*5-2)
plt.title("y=5Θ0-2")
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```
